```python
# Sujetos de la base de datos
subjects = list(np.arange(8,32))
# canales de la base de datos
channels = ['Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8', 'T3', 'C3', 'Cz', 'C4',
            'T4', 'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2']
n_channels = len(channels)
# Frecuencia de muestreo de la base de datos.
sampling_freq  = 1000
# ubicación de la base de datos.
load_path = '/content/drive/Shareddrives/GCPDS/databases/Databases_musica/Affective music listening 2/'

f_frec =  np.array([[1,4],[4,8],[8,12],[12,16],[16,20],[20,30],[30,45]])
#vtw= np.array([[0,2],[1,3],[2,4],[3,5],[4,6],[5,7]])
tau =  2
overlap=1-0.50
ti = 1.5
tf = 5.5
tti = np.arange(ti,tf-tau+(tau*overlap),tau*overlap)
ttf = np.arange(ti+tau,tf+tau*overlap,tau*overlap)
vtw = np.array([tti,ttf]).T
y_lab = loadmat('layer.mat')['lay']
#vtw= np.array([[2,6]])
#f_frec = np.array([[2.5,4.5]])
# subjects=np.delete(subjects,[28,33])
lab_ = np.arange(0,1240,40)
for sbj in range(31):#subjects[2:4]:
    # db = loaddb.GIGA('../input/bkernel-csp-lda/Giga/')
    # db.load_subject(sbj)
    # print(f'The subject {sbj} has {db.runs} runs')
    # Xraw, y = db.get_data(reject_bad_trials=True,classes=['left hand mi', 'right hand mi'])
    #run[i], class_[i] = db.get_run(i,classes=['left hand', 'right hand'])
    Xraw = loadmat(load_path + f"Sujeto_"+str(sbj+1)+".mat")['music']
    print('sujeto: ', sbj+1)
    Xraw = np.asarray(Xraw)
    Xraw=Xraw[:,:,::4]
    print(Xraw.shape)
    y = np.asarray(y_lab[:,lab_[sbj]:lab_[sbj]+40],dtype=np.float64).ravel()
    y[y==0]=-1
    print(y.shape)
    fs = sampling_freq/4
    #fs =db.metadata['sampling_rate']
    channels_names = channels
    montage = mne.channels.read_montage('standard_1005', channels_names)
    info = mne.create_info(channels_names, sfreq=fs, ch_types="eeg",
                            montage=montage) 
    feat = FB_feats(fs=fs,vtw=vtw,f_frec=f_frec,gammad=1,feat='kcon')
    t=time()
    xxx = feat.fit_transform(Xraw,y) 
    #print(time()-t)
    #opt = tf.keras.optimizers.RMSprop(learning_rate=1e-1)
    #%%%%%% 
    epochs=100
    batch_size=256
    optimizer='Adam'
    validation_split=0.1
    verbose=0
    gamma=1.0
    init_ = 'def_keras'
    learning_rate = 1e-3  
    steps=[('scal1',StandardScaler()),
        ('proy',Keras_CKA(epochs=epochs,learning_rate=learning_rate,batch_size=batch_size,optimizer=optimizer,
                    validation_split=validation_split,verbose=verbose,gamma=gamma,init_ = init_)),
           ('scal2',StandardScaler()),
            ('cla', SVC())]  
    method = Pipeline(steps,memory='/kaggle/tmp/datospipeline')
    parameters ={'proy__Q':[0.25,0.5,0.75,0.9],
                    'proy__l1_param':[0, 0.001, 0.1, 1],#0, 0.001, 0.1, 1
                    'proy__l2_param':[0, 0.001, 0.1, 1],
                    'cla__gamma': [1e-2,1e-1,1,1e1,1e2],
                    'cla__C': [1,1e1,1e2,1e3]}
    scores = {'acc': 'accuracy','kappa': make_scorer(cohen_kappa_score,greater_is_better=True)}    
    cv = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
    grid_search = GridSearchCV(method, parameters,cv=cv,verbose=0,
                               scoring=scores,refit='acc',n_jobs=-1)
    #grid_search = GridSearchCV(method, parameters,cv=cv,verbose=10,
     #                          scoring=kappa_,n_jobs=-1)
    #try:
    t=time()
    grid_search.fit(xxx,y)
    print(time()-t)
    results = grid_search.cv_results_
    #%% topoplots
    cka = grid_search.best_estimator_['proy']
    rho = abs(cka.A_out['Proj/kernel:0']).mean(1)
    savedata ={
        'steps':steps,
        'best_param':grid_search.best_params_,
        'rho':rho,        
        'cv_results':results

    }
    print(time()-t)
    # dump(savedata,'sujeto_'+str(sbj)+'_cka_featuresCSP_BCI2a_acc.joblib')
    # dump(savedata,'sujeto_'+str(sbj)+'_cka_featuresgauss_GIGA_MI_acc.joblib')
    dump(savedata,'sujeto_'+str(sbj)+'_cka_featuresgauss_Music_acc.joblib')
    feat.rho_plot(rho,info,figsize=(15,15),save=True,sbj=sbj,acc=np.round(grid_search.best_score_*100,2),format='eps',Ch=len(channels_names))
    print('sbj'+str(sbj)+'acc'+str(np.round(grid_search.best_score_*100,4)))
    #print(time()-t)
```
